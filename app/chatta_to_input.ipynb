```python
import pandas as pd
import numpy as np
import glob
import datetime
```

```python
dates_list = glob.glob("../data/data_Chatta/*")
```

```python
column_names = ['pickup_node', 'pickup_lon', 'pickup_lat', 'dropoff_node', 'dropoff_lon', 'dropoff_lat', 'time']
request_df = pd.DataFrame()
for date in dates_list:
    day_df = pd.read_csv(date+'/requests/requests.csv', names = column_names)
    date_str = date.split('/')[-1]
    month, day = date_str.split('-')
    date_str = datetime.datetime(2021, int(month), int(day), 0, 0, 0).date()
    day_df['date'] = date_str
    request_df = request_df.append(day_df)
```

```python
request_df = request_df.reset_index(drop=True)
```
